#### 16. Given a paragraph, find the most frequent word (case-insensitive, ignoring punctuation). (★★☆)

```python
from typing import Counter
paragraph = '''Python is a high-level, general-purpose programming language that emphasizes code
 readability, simplicity, and ease-of-writing with the use of significant indentation,[38] an 
 extensive ("batteries-included") standard library, and garbage collection. Python supports multiple
   programming paradigms but with an
 emphasis on object-oriented programming and dynamic typing.'''

count = Counter(paragraph.split())
print(count.most_common())
```
