## 4. Control Flow & Comprehensions[Comprehensions docs](https://docs.python.org/3/tutorial/datastructures.html#list-comprehensions)

## 6. Object-Oriented Programming[Classes docs](https://docs.python.org/3/tutorial/classes.html)

Classes bundle data (attributes) and behavior (methods) together. Python supports inheritance, special/dunder methods, and properties.**Worked example:** a simple class with a dunder method.

```python
class Point:    def __init__(self, x, y):        self.x, self.y = x, y    def __repr__(self):        return f'Point({self.x}, {self.y})'    def distance_to(self, other):        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5p1, p2 = Point(0, 0), Point(3, 4)print(p1, p2, p1.distance_to(p2))
```

## 7. Error Handling, File I/O & Iterators[Errors and exceptions docs](https://docs.python.org/3/tutorial/errors.html)

`try`/`except`/`finally` handles runtime errors gracefully. `with` statements manage resources like files. Generators produce values lazily using `yield`.**Worked example:** handling a specific exception.

```python
def safe_divide(a, b):    try:        return a / b    except ZeroDivisionError:        return float('inf')print(safe_divide(10, 2), safe_divide(10, 0))
```
